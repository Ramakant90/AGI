```python
def tokenize(sentence):
    return sentence.lower().split()

vocab = set()
for pair in pairs:
    for sentence in pair:
        vocab.update(tokenize(sentence))

vocab = sorted(vocab)
word2idx = {word: i+1 for i, word in enumerate(vocab)}  # +1 for padding
idx2word = {i+1: word for i, word in enumerate(vocab)}

vocab_size = len(vocab) + 1
print("Vocabulary Size:", vocab_size)
```

```python
def encode(sentence, max_len=8):
    tokens = tokenize(sentence)
    ids = [word2idx[w] for w in tokens if w in word2idx]
    ids = ids + [0]*(max_len - len(ids))
    return torch.tensor(ids[:max_len])

def decode(tensor):
    words = [idx2word.get(idx, "") for idx in tensor if idx != 0]
    return " ".join(words)
```

```python
class ChatBot(nn.Module):
    def __init__(self, vocab_size, hidden_size):
        super(ChatBot, self).__init__()
        self.embedding = nn.Embedding(vocab_size, hidden_size)
        self.rnn = nn.GRU(hidden_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, vocab_size)

    def forward(self, x):
        x = self.embedding(x)
        output, _ = self.rnn(x)
        out = self.fc(output)
        return out
```

```python
model = ChatBot(vocab_size, hidden_size=32)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

epochs = 5000

for epoch in range(epochs):
    total_loss = 0
    for inp, target in pairs:
        inp_ids = encode(inp)
        target_ids = encode(target)

        optimizer.zero_grad()
        output = model(inp_ids.unsqueeze(0))

        loss = criterion(output.view(-1, vocab_size), target_ids)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    if (epoch+1) % 100 == 0:
        print(f"Epoch {epoch+1}/{epochs}, Loss: {total_loss:.4f}")

print("âœ… Training complete!")
```

```python
def chat():
    print("ğŸ¤– RamakantBot ready! (type 'quit' to stop)\n")
    while True:
        msg = input("You: ")
        if msg.lower() == "quit":
            break
        inp = encode(msg)
        output = model(inp.unsqueeze(0))
        pred = torch.argmax(output, dim=2)
        response = decode(pred.squeeze())
        print("Bot:", response)

chat()
```
